# RTC Category Network

# Insight

Community acts as the central hub connecting multiple categories.
Strong links between Career, Skills, and Internship indicate that skill development directly supports career opportunities.
Funding is connected to Programs, showing that financial resources enable educational initiatives.


# Key Takeaway

Community engagement is the foundation that drives career growth, skill development, and mentorship opportunities.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

# Categories from the dataset
G.add_nodes_from([
    "Community",
    "Recruiting",
    "Career",
    "Program",
    "Internship",
    "Events",
    "Skills",
    "Funding",
    "Mentorship"
])

# Example relationships
G.add_edges_from([
    ("Community", "Events"),
    ("Community", "Mentorship"),
    ("Community", "Skills"),
    ("Recruiting", "Career"),
    ("Recruiting", "Internship"),
    ("Career", "Skills"),
    ("Career", "Program"),
    ("Program", "Funding"),
    ("Program", "Mentorship"),
    ("Internship", "Career"),
    ("Events", "Skills")
])

plt.figure(figsize=(8, 6))

nx.draw(
    G,
    with_labels=True,
    node_size=3500,
    node_color="red",
    font_size=10,
    font_weight="bold"
)

plt.title("RTC Business Metrics Network")

plt.show()
```

# Degree Centrality Analysis: Identifying the Most Influential Categories

# Insight

Degree centrality measures how many direct connections each category has within the RTC ecosystem. Community and Career have the highest centrality scores, making them the most influential nodes in the network.

# Key Takeaway

The RTC ecosystem depends heavily on Community engagement and Career development. These categories act as the primary connectors that influence multiple organizational outcomes.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_edges_from([
    ("Community", "Career"),
    ("Community", "Events"),
    ("Community", "Mentorship"),
    ("Career", "Skills"),
    ("Career", "Internship"),
    ("Internship", "Recruiting"),
    ("Program", "Funding"),
    ("Program", "Skills")
])

centrality = nx.degree_centrality(G)

nodes = list(centrality.keys())
scores = list(centrality.values())

plt.figure(figsize=(8,5))

plt.bar(
    nodes,
    scores,
    color=[
        "#7F0000",
        "#B22222",
        "#DC143C",
        "#FF4500",
        "#FF6F00",
        "#FF8C00",
        "#FFA500",
        "#FFD700"
    ]
)

plt.title("Degree Centrality of RTC Categories")

plt.xlabel("Category")
plt.ylabel("Centrality Score")

plt.xticks(rotation=45)

plt.show()
```

Using NetworkX, I modeled RTC business metrics as a graph network to identify key ecosystem relationships. Community emerged as the central hub, while Career functioned as a bridge connecting skill development initiatives to employment outcomes. The analysis demonstrates how graph theory can be applied to business and community data to uncover influential entities and strategic relationships.
