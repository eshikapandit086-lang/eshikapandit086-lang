```python

import pandas as pd
df = pd.read_csv('rtc_master_dataset.csv')
```

```python
print("=" * 40)
print("TASK 1 — DATA QUALITY ASSESSMENT")
print("=" * 40)

print(f"\nTotal Records : {len(df)}")
print(f"Null Values   : {df.isnull().sum().sum()}")
print(f"Duplicate Rows: {df.duplicated().sum()}")
print(f"Empty Pages   : {(df['text'].str.strip() == '').sum()}")

print("\n--- Text Length Distribution ---")
print(df['text'].str.len().describe().round(1))

print("\n--- Report Distribution ---")
print(df['report_name'].value_counts())
```

```python
df['text'] = (df['text']
    .str.replace(r'[\n\r\t]+', ' ', regex=True)
    .str.replace(r' {2,}', ' ', regex=True)
    .str.strip()
)
print("✓ Whitespace cleaned")
```

```python
df['report_name_clean'] = (df['report_name']
    .str.replace('-', '_', regex=False)
    .str.strip()
)
print(df['report_name_clean'].value_counts())
```

```python
print(f"Duplicate Rows : {df.duplicated().sum()}")
print(f"Null Values    : {df.isnull().sum().sum()}")
print(f"Unique Reports : {df['report_name_clean'].nunique()}")
```

```python
df['text_length'] = df['text'].str.len()
df['word_count']  = df['text'].str.split().str.len()
print("✓ Columns added")
```

```python
df.groupby('report_name_clean')['word_count'].sum()
```

```python
df['word_count'].describe().round(1)
```

```python
import matplotlib.pyplot as plt
```

```python
df['report_name_clean'].value_counts().plot(kind='bar', color='#E8651A')
plt.title("Pages per Report")
plt.show()
```

```python
df['word_count'].plot(kind='hist', color='#F4A460')
plt.title("Word Count Distribution")
plt.show()
```

```python
top5 = df.nlargest(5, 'word_count')[['report_name_clean', 'word_count']]
bottom5 = df.nsmallest(5, 'word_count')[['report_name_clean', 'word_count']]

pd.concat([top5, bottom5]).plot(kind='bar', x='report_name_clean', y='word_count', color='#FF8C00')
plt.title("Longest vs Shortest Pages")
plt.show()
```

```python
df.boxplot(column='text_length', by='report_name_clean', figsize=(20,10),
           boxprops=dict(color='#E8651A'),
           medianprops=dict(color='#FF4500'),
           whiskerprops=dict(color='#F4A460'),
           capprops=dict(color='#F4A460'))
plt.title("Text Length by Report")
plt.show()
```

```python
import numpy as np

# create page buckets
df['page_bucket'] = pd.cut(df['page_number'], bins=5, labels=['1-6','7-12','13-18','19-24','25+'])

pivot = df.pivot_table(index='report_name_clean', columns='page_bucket', values='word_count', aggfunc='mean').round(0)

plt.figure(figsize=(10, 5))
plt.imshow(pivot, cmap='Oranges', aspect='auto')
plt.colorbar(label='Avg Word Count')
plt.xticks(range(len(pivot.columns)), pivot.columns)
plt.yticks(range(len(pivot.index)), [r[-4:] for r in pivot.index])
plt.title("Content Density Heatmap — Report vs Page Range")
plt.xlabel("Page Range")
plt.ylabel("Report Year")
plt.tight_layout()
plt.show()
```

```python
df.groupby('report_name_clean')['word_count'].mean().plot(kind='barh', color='#FF8C00')
plt.title("Avg Word Count per Report")
plt.xlabel("Avg Words")
plt.show()
```

```python
df.groupby('report_name_clean')['word_count'].sum().plot(kind='bar', color='#E8651A')
plt.title("Total Words per Report")
plt.show()
```

```python
df['word_count'].plot(kind='line', color='#FF4500')
plt.title("Word Count per Page")
plt.xlabel("Page Index")
plt.ylabel("Word Count")
plt.show()
```
